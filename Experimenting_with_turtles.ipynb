```
import ColabTurtle
from ColabTurtle.Turtle import *
```

```
alex = ColabTurtle.Turtle
alex.initializeTurtle()
theta=120
for l in range(70):
  alex.left(120)
  alex.forward(100)
  alex.left(120)
  alex.forward(100)
  alex.left(120)
  alex.forward(100)
  l+=5
  alex.left(5)
```

```
alex = ColabTurtle.Turtle
alex.initializeTurtle()
theta=120
len=100
angle = 360-360/5
for l in range(70):
  alex.left(angle)
  alex.forward(len)
  alex.left(angle)
  alex.forward(len)
  alex.left(angle)
  alex.forward(len)
  alex.left(angle)
  alex.forward(len)  
  alex.left(angle)
  alex.forward(len)  
  alex.left(30)
  if l%20==0:
    len+=10
```

```
karev = ColabTurtle.Turtle
karev.initializeTurtle()
theta=120
len=100
angle = 120  #360-360/3
for l in range(0,70,10):
  karev.penup()
  karev.left(angle)
  karev.forward(len)
  karev.pendown()
  karev.left(angle)
  karev.forward(len)
  karev.penup()
  karev.left(angle)
  karev.forward(len)
  karev.left(30)
  len+=5
```
